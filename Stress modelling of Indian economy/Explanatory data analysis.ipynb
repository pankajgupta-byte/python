# causal analysis

```python
import statsmodels.api as sm
```

```python
X = corr_mat[['gdp_growth','Fiscal Deficit','Inflation_rate','Repo Rate ','credit_growth_in_percent','crisis']]  
Y = corr_mat['BSE']  
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
summary = model.summary()
print(summary)
```

```python
final_data = indicators.drop(columns=['Date','Imports','Trade Balance','credit','forex','CPI','wage_rate','Exports','Exchange rate','IIP_index','Crude_Price', 'Debt_percentage','BCI '])
final_data
```

```python
print(final_data.isna().sum())
```

```python
final_data_clean = final_data.dropna()
p_value_matrix = pd.DataFrame(np.zeros((len(final_data_clean.columns), len(final_data_clean.columns))),
                              columns=final_data_clean.columns, index=final_data_clean.columns)
for BSE in final_data_clean.columns:
    for gdp_growth in final_data_clean.columns:
        if BSE != gdp_growth:  
            X = final_data_clean[gdp_growth]  
            Y = final_data_clean[BSE] 
            X = sm.add_constant(X)  
            model = sm.OLS(Y, X).fit()  
            p_value_matrix.loc[BSE, gdp_growth] = model.pvalues[1]  
print(p_value_matrix)
```
