```python
a1 = pd.merge(bse, gdp_growth, on='Date', how='left')
a1
```

```python
a2 = pd.merge(a1, fiscal_deficit, on='Date', how='left')
a2.shape
```

```python
a2 = pd.merge(a1, fiscal_deficit, on='Date', how='left')
a3 = pd.merge(a2,inflation_growth, on='Date', how='left')
a4 = pd.merge(a3, exchange_rate, on='Date', how='left')
a5 = pd.merge(a4, trade_bal, on='Date', how='left')
a6 = pd.merge(a5, crude_oil_price, on='Date', how='left')
a7 = pd.merge(a6, debt_to_percentage, on='Date', how='left')
a8 = pd.merge(a7, bci, on='Date', how='left')
a9 = pd.merge(a8, repo_rate, on='Date', how='left')
a10 = pd.merge(a9, forex, on='Date', how='left')
a11 = pd.merge(a10, iip, on='Date', how='left')
a12 = pd.merge(a11, credit_growth, on='Date', how='left')
```

```python
a13 = pd.merge(a12, cpi_index, on='Date', how='left')
a14 = pd.merge(a13, wage_rate, on='Date', how='left')
indicators = pd.merge(a14, crisis, on='Date', how='left')
```
