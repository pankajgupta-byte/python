# data preprocessing

```python
pop['Airing_End'] = pop['Airing_End'].replace("Still Airing", "2025-02-24")
```

```python
rev["Production Budget"] = rev["Production Budget"].astype(str).str.replace(r"\D", "", regex=True).astype('int64')
```

```python
rev["Streaming Cost"] = (rev["Streaming Cost"].astype(str).str.replace(r"\D", "", regex=True).replace("", "0")  .astype("Int64"))
rev["Merchandise revenue"] = (rev["Merchandise revenue"].astype(str).str.replace(r"\D", "", regex=True).replace("", "0").astype("Int64"))
rev["Streaming revenues"] = (rev["Streaming revenues"].astype(str).str.replace(r"\D", "", regex=True).replace("", "0").astype("Int64"))
```

```python
rev["Total revenues"] = (rev["Total revenues"].astype(str).str.replace(r"\D", "", regex=True).replace("", "0").astype("Int64"))
```

```python
import re
```

```python
df1 = pd.DataFrame(df['Genre'])
genre_list = [
    "Drama", "Comedy", "Action", "Adventure", "Romance", "Fantasy", "Sci-Fi",
    "Supernatural", "Mystery", "Horror", "Slice of Life"
]
def extract_genres(text):
    pattern = r'(' + '|'.join(sorted(genre_list, key=len, reverse=True)) + r')'
    return re.findall(pattern, text)

df1["Extracted_Genres"] = df1["Genre"].apply(extract_genres)

print(df1[["Genre", "Extracted_Genres"]])
```

```python
df["Extracted_Genres"] = df1["Extracted_Genres"].apply(lambda x: ", ".join(x))
```

```python
df['genre_count']=df["Extracted_Genres"].str.split(", ").apply(len)
```

```python
df['aired_days']=df['aired_days'].dt.days.astype(int)
```

```python
df['release_frequency']=(df["aired_days"] / df["Episodes"]).round()
```

```python
df['watch_hour_per_member']=df['Watch Hours']/df['Members']
```

```python
df['engagement_index']=(df['User_Reviews']/df['Favorites']).sort_values().tail(1200)
```

```python
df2 = pd.DataFrame(df["Streaming Platform_y"])
streaming_platforms = [
    "Netflix", "Amazon Prime Video", "Disney+", "Hulu", "HBO Max", "Apple TV+",
    "YouTube", "Peacock", "Paramount+", "Crunchyroll", "Funimation", "Tubi",
    "Pluto TV", "Vudu", "Sling TV", "fuboTV", "Rakuten Viki", "BBC iPlayer",
    "Showtime", "Starz", "Shudder", "Acorn TV", "Kanopy", "Mubi", "BritBox"
]

def extract_platforms(text):
    pattern = r'(' + '|'.join(sorted(streaming_platforms, key=len, reverse=True)) + r')'
    return re.findall(pattern, text)

df2["Extracted_platforms"] = df2["Streaming Platform_y"].apply(extract_platforms)

print(df2[["Streaming Platform_y", "Extracted_platforms"]])
```

```python
df['Extracted_platforms']=df2["Extracted_platforms"].apply(lambda x: ", ".join(x))
```

```python
df["Extracted_platforms"]=-df["Extracted_platforms"].str.split(", ").apply(len)
```
